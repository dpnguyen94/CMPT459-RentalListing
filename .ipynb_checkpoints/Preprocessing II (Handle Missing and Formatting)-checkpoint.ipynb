```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
trainDataDir = "dataset/two-sigma-connect-rental-listing-inquiries/trainNaN.json.zip"
```

```python
trainData = pd.read_json(trainDataDir, convert_dates=['created'])
trainData = trainData.reset_index(drop=True)
```

# Find missing building id

```python
#https://stackoverflow.com/questions/27159189/find-empty-or-nan-entry-in-pandas-dataframe
buildingNAN = trainData.loc[trainData['building_id'].isnull()]
buildingNAN.head(4)
```

```python
trainDataNonNAN = trainData.dropna(subset=['building_id'])
trainDataNonNAN['building_id'].isnull().any()
```

```python
buildingID = trainDataNonNAN[['building_id','latitude','longitude']].dropna()
buildingID.groupby(['latitude','longitude']).count()['building_id'].unique()
```

```python
a = buildingNAN.merge(buildingID, how='inner', on=['latitude','longitude'])
a.head()
```

```python
a = a.drop_duplicates(subset=['listing_id']).drop('building_id_x',axis=1).rename(columns ={'building_id_y':'building_id'})
a
```

```python
trainData = a.append(trainData,ignore_index = True)
trainData
```

```python
trainData['listing_id'].value_counts()
```

## Extract address

```python
address = trainData[['building_id','latitude','longitude','street_address']].dropna()
address
```

```python
#https://stackoverflow.com/questions/57503786/python-replace-abbreviation-in-text
replacers = {
    'st': 'street',
    'e': 'east',
    'w': 'west',
    's': 'south',
    'n': 'North',
}
address['street_address'] = (
    address['street_address']
    .str.lower()
    .str.replace('[.,]','')
    .str.split()
    .apply(lambda x: ' '.join([replacers.get(e, e) for e in x]))
)
address
```

```python
address.groupby(['latitude','longitude','street_address']).count()
```
